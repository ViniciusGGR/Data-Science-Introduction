# Ambientes Virtuais

- **Ambientes virtuais**, são *ambientes Python* de **área restrita** com *versões específicas de bibliotecas* (ou, dependendo da configuração, até mesmo do Python).

- Um **ambiente virtual** é uma "*cópia independente*" de tudo o que é *necessário* para **executar o programa**. Incluindo o "intérprete" do **Python** e *todas as bibliotecas* de que o programa precisa.
  - Usar um **ambiente virtual** garante que o programa terá *acesso às versões e aos recursos corretos* para ser executado corretamente.

### Criando um Ambiente Virtual



Para criar um **ambiente virtual** chamando o *módulo* `venv`. O *módulo* espera um **nome** como um **argumento**.

- Vá para o **diretório** onde deseja manter o projeto.
- Use o *comando* a seguir para *chamar o módulo* `venv`:
  
  ```
  python -m venv env
  ```

  - Alguns **diretórios** seram criados. Os *nomes* de **diretório** diferem dependendo do sistema operacional (SO).
  - **Diretórios** no Windows:

    ```
    /env
      /include
      /lib
      /Scripts
    ```
  - **Diretórios** no Linux e macOS:

    ```
    /env
      /bin
      /include
      /lib
    ```

- O *ambiente* precisa do **diretório** `env` para manter o "*controle dos detalhes*", como qual **versão do Python** e quais **bibliotecas** estão sendo utilizadas.
- Não coloque os *arquivos de programa* no **diretório** `env`. Coloque os *arquivos de programa* no **diretório** `src` ou algo semelhante. A *estrutura do projeto* pode ter a seguinte aparência:

  ```
  /env
  /src
    program.py  
  ```
